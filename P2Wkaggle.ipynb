```python
data_text['s1'] = data_text['s1'].str.replace('[', '')
data_text['s1'] = data_text['s1'].str.replace(']', '')
data_text[['s_1', 's_2']] = data_text['s1'].str.split(",",1,expand=True)
data_text['s_1'] = data_text['s_1'].str.replace("'", '')
data_text['s_2'] = data_text['s_2'].str.replace("'", '')
```

```python
data_text.head()
```

```python
data_text['s2'] = data_text['s2'].str.replace('[', '')
data_text['s2'] = data_text['s2'].str.replace(']', '')
data_text[['s_21', 's_22']] = data_text['s2'].str.split(",",1,expand=True)
data_text['s_21'] = data_text['s_21'].str.replace("'", '')
data_text['s_22'] = data_text['s_22'].str.replace("'", '')
```

```python
s1_embbeding = []
for s1,s2 in zip(data_text['s_1'],data_text['s_2']):
  if s1 and s2:
    embeddings =  model.encode([s1,s2])
    s1_embbeding.append(np.mean(embeddings,axis = 0))
  elif s1:
    s1_embbeding.append(model.encode(s1))  
  elif s2:
    s1_embbeding.append(model.encode(s2))
  else:
    s1_embbeding.append(0)
```

```python
s2_embbeding = []
for s1,s2 in zip(data_text['s_21'],data_text['s_22']):
  if s1 and s2:
    embeddings =  model.encode([s1,s2])
    s2_embbeding.append(np.mean(embeddings,axis = 0))
  elif s1:
    s2_embbeding.append(model.encode(s1))  
  elif s2:
    s2_embbeding.append(model.encode(s2))
  else:
    s2_embbeding.append(0)
```

```python
from typing import Iterable 
#from collections import Iterable                            # < py38


def flatten(items):
    """Yield items from any nested iterable; see Reference."""
    for x in items:
        if isinstance(x, Iterable) and not isinstance(x, (str, bytes)):
            for sub_x in flatten(x):
                yield sub_x
        else:
            yield x

pred = flatten(clf.predict(X_test).tolist())
```
